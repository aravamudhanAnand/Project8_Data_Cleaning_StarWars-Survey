```python
import pandas as pd
star_wars= pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

ID_notnull= pd.notnull(star_wars["RespondentID"])
star_wars= star_wars.loc[ID_notnull,:]
                                                   
                                                   
                                                   
```

# Cleaning and Mapping Yes/No Columns

```python
yes_no = {
    "Yes": True,
    "No": False
}
col_names= ["Have you seen any of the 6 films in the Star Wars franchise?","Do you consider yourself to be a fan of the Star Wars film franchise?"]

for row in col_names:
    star_wars[row]= star_wars[row].map(yes_no)
star_wars.head()
    
```

# Cleaning and Mapping Checkbox Columns

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    

```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

# Cleaning the Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars.loc[:,star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

# Finding the Highest Ranked Movie

```python
ranking_means=star_wars.loc[:,star_wars.columns[9:15]].mean()
print(ranking_means)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

ranking_means.plot.bar()
plt.show()
```

# Finding the Most Viewed Movie

```python
most_seen=star_wars.loc[:,star_wars.columns[3:9]].sum()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

most_seen.plot.bar()
plt.show()
```

# Exploring the Data by Binary Segments

```python
males = star_wars.loc[star_wars["Gender"] == "Male",:]
females = star_wars.loc[star_wars["Gender"] == "Female",:]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```
